# Exploratory Data Analysis (EDA)
 
  ### 1. Finding the Missing Values or Nan Values.

```python
data.isna().sum()
```

* so there are total 517 Nan Values in the column called <span style= 'color: Red; font-size:20px'> Precip Type. </span>

### So Now  we should remove that missing values / Nan  values or fill their values.

```python
data['Precip Type'].unique()
```

```python
data['Summary'].unique()
```

* As we know that this column doesn't give us that much information how much we want so we will use the fill na method  to fill that Nan Values with some values.
* Therefore, there is the method to fill this rows by some values.
* So i used the method called 'bfill' --> "backwardfill" and there is one more method called the 'ffill'--> 'forwardfill'

* <span style ="color :Red; font-size: 20px;"> .fillna(method ='bfill')</span>

```python
clean_data=data.fillna(method='bfill') #--> and we have create a new variable in that variable we have store that dataset with no missing values 
```

* Now there is no missing values and we have filled everything with some values.

```python
clean_data.isna().sum() #--> No missing values.
```

### As we know that there was the column called the "Formatted Date" in the dataset but the dtype of that column was an object to convert it into the the datetime format we use this:
* pd.to_datetime()-----> where we should keep the parameter <span style="font-size:20px;font-weight: bold;">utc =True</span>

```python
clean_data['Formatted Date']=pd.to_datetime(clean_data['Formatted Date'],infer_datetime_format=True, utc=True)
```

```python
clean_data['Formatted Date']
```

###  There is an method where we  can find the sum of all unique values in the dataset
* .nunique()

```python
clean_data.nunique()
```

```python
clean_data.sample(20)
```

* Now see the Datatype shown below, the column "Formatted Date" datatype is now changed to the datetime64[ns,utc]

```python
clean_data.dtypes 
```

```python
clean_data=clean_data.set_index('Formatted Date')
clean_data
```

* As our main work of the dataset is done so now we will move towards the resampling our data into the monthly format so that we can extract the data on the basis of the monthly format.
* **we can resample our data by using this the method** **.resample('[Rule]')**

```python
# Now to Convert Our Data to montly format from Hourly. 
data_monthly=clean_data.resample('MS').mean()
data_monthly.head(10)
```

* **Extracting the data from April 2006 to April 2016 for the column Apparent Temperature (C) and Humidity. so that we can perform some Graphical Representation**

```python
data_target=data_monthly.loc['2006-04-01':'2016-4-01', ['Apparent Temperature (C)', 'Humidity']]
```

```python
data_target #--> this is average Apparent Temperature and Humidity on the monthly basis factor.
```

# Data Visualization 

### First of all let's see the Trends of the Apparent Temperature (C) and Humidity by using the Line Plots.
## Line Charts

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,8))
plt.plot(data_target.index, data_target['Apparent Temperature (C)'],lw=5, mfc= 'yellow')
plt.plot(data_target.index, data_target["Humidity"], lw=5)
plt.xlabel("Year")
plt.title("Trends of Apparent Temperature and Humidity")
plt.ylabel("Apparent Temperature (C)")
plt.legend(["Apparent Temperature", "Humidity"]);
```

* **by observing this we knew that the "Apparent Temperature (C)" is increasing and the mid of the year decreasing this trends says that there is slightly increment in the month of January, february, March, April and May.** 
<br>
* **And the Humidity there is no increment in past 10 years it is approximately same.**

### Finding the Correlation between the two variables

```python
data_target.corr() #---> lets see what is the Relation between the Apparent Temperature and Humidity. by the method 
# called corr() which shows the Correlation.
```

```python
plt.title("Heatmap Relationship")
sns.heatmap(data_target.corr(), annot =True ,cmap="Greens");
```

## Now lets have some Scatter Plots so that we can get the relationship between this the Two variable.

```python
plt.figure(figsize=(15,8))
plt.title("Apparent Temperature v/s Humidity")
sns.scatterplot('Apparent Temperature (C)', 'Humidity', data= data_target, s=100);
```

* **by observing this plot we understood that there is some relationship between the Apparent Temperature (C) and Hummidity but we are not getting that so we plot it by the linear Regression.**

```python
plt.figure(figsize=(15,8))
sns.lmplot("Apparent Temperature (C)", "Humidity", data=data_target);
```

* **so by this plot there is some negative relationhip between the Apparent Temperature and Humidity that the regression line passes negatively and there are some outliers which doesnt matter in the plots.**

```python
# Lets see some relationship between some other variables.
clean_data[['Apparent Temperature (C)', "Visibility (km)"]].corr()
#There is some positive correlation between this variable called Apparent Temperature and Visibility
```
